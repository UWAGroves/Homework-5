```python
#Import packages and link Gdrive

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import expon
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss

from google.colab import drive

drive.mount('/content/drive')

filedir = "/content/drive/MyDrive/"
```

```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creates plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, linestyle='--', color='blue', label='sin(X)')

# Adds labels
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(X)')

# Show the plot
plt.legend()
plt.show()
```

```python
# Load data
tips = sns.load_dataset("tips")

# Creates plot
plt.figure()
sns.histplot(tips["total_bill"], bins=20, kde=True, color='green')

# Adds labels
plt.title("Histogram with KDE of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
# Creates scatter plot with regression lines
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", palette="Set1", markers=[".", "+"], aspect=1.2)

# AddS title
plt.title("Total Bill vs. Tip by Smoker Status")

# Show the plot
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = tips.corr(numeric_only=True)

# Creates heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, center=0)

# Add title
plt.title("Correlation Matrix of Tips Dataset")

# Show the plot
plt.show()
```

```python
# Creates pairplot
sns.pairplot(tips, hue="time", diag_kind="kde")

# Show the plot
plt.show()
```
